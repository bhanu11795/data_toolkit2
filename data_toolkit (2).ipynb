question 2-How does broadcasting work in NumPy?

  - Broadcasting refers to the way NumPy handles element-wise operations on arrays of different shapes. When two arrays have different shapes, NumPy automatically adjusts their shapes to make them compatible for element-wise operations. For example, if you add a 2D array with a 1D array, the 1D array is broadcasted (repeated) across the rows of the 2D array to perform the operation. This eliminates the need for explicit loops and leads to more efficient computation.

question 6-What are the differences between NumPy arrays and Python lists?

 - NumPy arrays are optimized for numerical operations. They support vectorization, meaning that operations can be applied to entire arrays without explicit loops. NumPy arrays are also homogeneous (i.e., all elements must have the same data type), which allows for efficient memory usage and faster computations. On the other hand, Python lists are heterogeneous (i.e., they can contain elements of different types) and are general-purpose containers, which makes them slower and less memory-efficient for numerical tasks.

question 8-What does the term “vectorized operation” mean in NumPy?

   - A vectorized operation in NumPy means applying an operation to an entire array (or a part of it) without using loops. For example, if you want to add 10 to every element of an array, instead of looping through each element, you can simply write arr + 10. NumPy internally performs these operations much faster than Python loops because it uses optimized C code under the hood, which can lead to significant performance improvements for large datasets.

question 21-Why is NumPy’s array slicing faster than Python’s list slicing?

   - NumPy’s array slicing is faster because NumPy arrays are implemented in contiguous memory blocks. When you slice an array, it creates a view into the original array, rather than copying the data. In contrast, Python lists are more general-purpose and slicing them requires copying the data into a new list, which is slower.
